```python
import numpy as np
import scipy as sp
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
%matplotlib inline
%precision 3
import japanize_matplotlib
japanize_matplotlib.japanize()
```

## 問題
あるソシャゲのガシャのSSRキャラが出る確率が0.05だとする。
### 問1
このガシャがP=0.05の幾何分布に従うと仮定して、x=SSRが出るまでのガシャ回数, y=確率の確率質量関数を描画しなさい。
### 問2
P=0.05の累積分布関数のグラフを描画し、10回ガシャを引いた時にSSRが当たる確率（期待値）はいくらか求めなさい。

## 解答
### 問1
このガシャがP=0.05の幾何分布に従うと仮定して、x=SSRが出るまでのガシャ回数, y=確率の確率質量関数を描画しなさい。

```python
# ガシャを引く回数の数列を生成
nums = np.arange(1,101,1)

# 生成した数列(確率変数)に対応する確率を生成
geom_pmf = sp.stats.geom.pmf(k=nums, p=0.05)
```

```python
# 確率質量関数を曲線＆棒グラフで可視化
plt.plot(nums, geom_pmf, color='orange')
plt.bar(nums, geom_pmf, color='green')
plt.xlabel('SSRが出るまでの回数', fontsize=13)
plt.ylabel('確率', fontsize=13)
plt.savefig('1.png')
```

## 問2
P=0.05の累積分布関数のグラフを描画し、10回ガシャを引いた時に5回以内でSSRが当たる確率（期待値）はいくらか求めなさい。

```python
# ガシャを引く回数の数列を生成
nums = np.arange(1,101,1)

# 生成した数列(確率変数)に対応する累積分布関数の生成
geom_cdf = sp.stats.geom.cdf(k=nums, p=0.05)

# 5回引いたときの累積密度
geom_under_10 = sp.stats.geom.cdf(k=10, p=0.05)
```

```python
# 可視化
plt.plot(nums, geom_cdf)
plt.plot(10, geom_under_10, 'ro')

plt.vlines(x=10, ymin=0, ymax=geom_under_10, linestyles='dashed', color='red')
plt.hlines(y=geom_under_10, xmin=0, xmax=10, linestyles='dashed', color='red')
plt.savefig('2.png')
```

```python
print('ガシャを10回引いた時、10回以内でSSRを引く確率：', geom_under_10)
```

## おまけ
SSR確率が1%の場合

```python
# ガシャを引く回数の数列を生成
nums = np.arange(1,101,1)

# 生成した数列(確率変数)に対応する累積分布関数の生成
geom_cdf = sp.stats.geom.cdf(k=nums, p=0.01)

# 10回引いたときの累積密度
geom_under_10 = sp.stats.geom.cdf(k=10, p=0.01)

# 可視化
plt.plot(nums, geom_cdf)
plt.plot(10, geom_under_10, 'ro')

plt.vlines(x=10, ymin=0, ymax=geom_under_10, linestyles='dashed', color='red')
plt.hlines(y=geom_under_10, xmin=0, xmax=10, linestyles='dashed', color='red')
plt.savefig('3.png')
```

```python
print('ガシャを10回引いた時、10回以内でSSRを引く確率：', geom_under_10)
```
