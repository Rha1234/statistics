```python
import numpy as np
import pandas as pd
import scipy as sp
from scipy import stats
from matplotlib import pyplot as plt
import seaborn as sns
sns.set()
%precision 3
%matplotlib inline
```

## 標本平均の分布を確認する
1. 平均4、標準偏差0.8の正規分布のデータを無作為に10000回発生させる（サンプルサイズは10, 標本数は10000）
2. それらの標本平均10000個の確率分布がどうなるか可視化する

```python
# 1.
# まずは同じパラメータの分布を使い回すためにRVオブジェクトを作成
test_norm = sp.stats.norm(loc=4, scale=0.8)

sample_mean = np.zeros(10000)

for i in range(10000):
    samples = test_norm.rvs(10)
    sample_mean[i] = np.mean(samples)
```

```python
# 2.
sns.histplot(data=sample_mean, kde=True, color='red')
plt.savefig('1.png')
```

```python
np.mean(sample_mean), np.std(sample_mean, ddof=1)
```

平均4標準偏差0.8の正規分布に近づいているように思える。しかし、標準偏差は母標準偏差より小さくなっている。

## 標本平均の母平均への収束を確認する
大数の法則によると、サンプルサイズを大きくすれば標本平均は母平均へ収束していくはず
1. サンプルサイズの配列を作成する
2. サンプルサイズごとに、平均4,標準偏差0.8の正規分布のデータを無作為に発生させ、標本平均を求める
3. サンプルサイズをX、標本平均をYとした折れ線グラフを描画して、標本平均の母平均（4）への収束を確認する

```python
# 1.
sample_sizes = np.arange(start=10, stop=100100, step=100)
samples = np.zeros(len(sample_sizes))
```

```python
# 2.
test_norm = sp.stats.norm(loc=4, scale=0.8)

np.random.seed(1)
for i in range(0, len(samples)):
    sample = test_norm.rvs(size=sample_sizes[i])
    samples[i] = np.mean(sample)
```

```python
# 3.
plt.plot(sample_sizes, samples, color="orange")
plt.xlabel('sample size')
plt.ylabel('sample mean')
plt.savefig('2.png')
```

グラフを見ると、サンプルサイズが大きくなればなるほど標本平均は母平均に収束していくことがわかる。

## サンプルサイズごとの標本平均の分布を確認する
1. サンプルサイズごとに、標本平均を計算する
2. それぞれの標本平均の分布（標本分布）を確認する

```python
import scipy as sp
test_norm = sp.stats.norm(loc=4, scale=0.8)

# 1.
def calc_sample_mean(size:int, n:int):
    sample_means = np.zeros(n)
    for i in range(n):
        sample = test_norm.rvs(size=size)
        sample_means[i] =  np.mean(sample)
    return sample_means
```

```python
# 2
ret_df = pd.DataFrame() # 結果保持用のdf

for i in range(10, 210, 50): # サンプルサイズ毎の標本分布を保持
    tmp_df = pd.DataFrame({
        "sample_mean": calc_sample_mean(i, 10000),
        "size": np.tile(f"size: {i}", 10000)
    })
    ret_df = pd.concat([ret_df, tmp_df])

sns.violinplot(data=ret_df, x='size', y='sample_mean')
plt.savefig('3.png')
```

同様に標本平均の標準偏差の分布を見ていく。

```python
size = np.arange(start=2,stop=102,step=2)
sample_mean_std = np.zeros(len(size))

np.random.seed(1)
for i in range(len(size)):
    sample_mean = calc_sample_mean(size=size[i],n=100)
    sample_mean_std[i] = np.std(sample_mean, ddof=1)

plt.plot(size, sample_mean_std, color='green')
plt.xlabel('sample size')
plt.ylabel('sample mean std')
plt.savefig('4.png')
```

グラフから、標本平均の標準偏差が小さくなっていくことがわかる。サンプルサイズを大きくすると信頼の出来る標本平均が得られることがわかった。

## 標本誤差
* 「標本平均の標準偏差」は数式で計算出せる。この数式で計算できるものを標本誤差という
* 標準誤差 = σ/N^(1/2)
